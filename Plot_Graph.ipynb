```
import networkx as nx 
import matplotlib as mpl
import matplotlib.pyplot as plt 
import numpy as np
```

```
vertices = int(input())
edges = int(input())
adjacent = [[0 for i in range(vertices)] for j in range(vertices)]

for i in range(edges):
  edge1 = int(input())
  edge2 = int(input())
  adjacent[edge1][edge2] = 1

rows=vertices
columns=vertices
```

```
#Undirected graph
G = nx.Graph() 
for i in range(rows): 
 for j in range( columns): 
   if adjacent[i][j] == 1 :
      G.add_edge(i,j)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos,node_size = 500)
nx.draw_networkx_edges(G, pos, arrows=False)
nx.draw_networkx_labels(G, pos)
plt.show() 
```

```
#Directional graph
G = nx.DiGraph() 
for i in range(rows): 
 for j in range( columns): 
   if adjacent[i][j] == 1 :
      G.add_edge(i,j)
 
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos,  node_size = 500)
nx.draw_networkx_edges(G, pos, arrows=True)
nx.draw_networkx_labels(G, pos)
plt.show()
```
